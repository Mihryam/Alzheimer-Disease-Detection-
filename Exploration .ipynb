```python
plt.hist(Demented['RHIPPOC'], bins=30)
plt.title('Right Hippocampus Distribution of Demented Patients')
plt.xlabel('Right Hippocampus')
plt.ylabel('Frequency')
plt.show()
```

# Checcking for Correlation Between Features
